```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("s3a://projeto-pi-datalake-raw/24-10-2022-lockdown-generate.csv",sep=";")
```

```python
df.columns = df.columns.str.lower()
```

```python
df.columns = df.columns.str.replace(" ", "_",regex=True)
```

```python
df.head()
```

```python
df.columns = df.columns.str.replace("id_", "id")
df.columns = df.columns.str.replace("mother/father_", "pais_autistas")
df.columns = df.columns.str.replace("child_age_", "idade")
df.columns = df.columns.str.replace("child_sex_", "sexo")
df.columns = df.columns.str.replace("asd_diagnosis_", "nivel_autismo")
df.columns = df.columns.str.replace("language_competences_", "nivel_comunicacao")
df.columns = df.columns.str.replace("cognitive_functioning", "funcoes_cognitivas")
df.columns = df.columns.str.replace("language_and_communication_", "status_comunicacao")
df.columns = df.columns.str.replace("emotional_regulation_", "controle_emocional")
df.columns = df.columns.str.replace("social_interaction_", "interacao_social")
df.columns = df.columns.str.replace("stereotypes_", "comportamento_esteriotipado")
df.columns = df.columns.str.replace("behavioral_problems_", "comportamento")
df.columns = df.columns.str.replace("restricted_interests_", "flexibilidade_interesses")
df.columns = df.columns.str.replace("autonomies_", "autonomia")
df.columns = df.columns.str.replace("tv", "frequencia_tv")
df.columns = df.columns.str.replace("video-games", "frequencia_video-games")
df.columns = df.columns.str.replace("play_with_frinds_online_", "joga_com_amigos_online")
df.columns = df.columns.str.replace("social_networks_", "redes_sociais")
```

```python
df = df.set_index('id')
```

```python
df.head()
```

```python
df["pais_autistas"].replace({1.0: "mae", 2.0: "pai", np.nan: "não informado"}, inplace=True)
df["sexo"].replace({0: "M", 1: "F", np.nan: "Não informado"}, inplace=True)
df["nivel_autismo"].replace({1: "leve", 2: "moderado", 3: "grave", np.nan: "não informado"}, inplace=True)
df["nivel_comunicacao"].replace({0: "sem_comunicacao", 1: "palavras_simples", 2: "frases_curtas", 3: "frases_complexas", np.nan: "não informado"}, inplace=True)
df["funcoes_cognitivas"].replace({1: "baixa", 2: "media", 3: "alta", np.nan: "não informado"}, inplace=True)
df["status_comunicacao"].replace({1: "melhorou", 2: "estavel", 3: "piorou", np.nan: "não informado"}, inplace=True)
df["controle_emocional"].replace({1: "melhorou", 2: "estavel", 3: "piorou", np.nan: "não informado"}, inplace=True)
df["interacao_social"].replace({1: "melhorou", 2: "estavel", 3: "piorou", np.nan: "não informado"}, inplace=True)
df["comportamento_esteriotipado"].replace({1: "melhorou", 2: "estavel", 3: "piorou", np.nan: "não informado"}, inplace=True)
df["comportamento"].replace({1: "melhorou", 2: "estavel", 3: "piorou", np.nan: "não informado"}, inplace=True)
df["flexibilidade_interesses"].replace({1: "melhorou", 2: "estavel", 3: "piorou", np.nan: "não informado"}, inplace=True)
df["autonomia"].replace({1: "melhorou", 2: "estavel", 3: "piorou", np.nan: "não informado"}, inplace=True)
df["frequencia_tv"].replace({1: "nunca", 2: "eventualmente", 3: "frequentemente", 0: "não informado"}, inplace=True)
df["frequencia_video-games"].replace({0: "nunca", 1: "eventualmente", 2: "frequentemente", np.nan: "não informado"}, inplace=True)
df["joga_com_amigos_online"].replace({0: "nunca", 1: "eventualmente", 2: "frequentemente", np.nan: "não informado"}, inplace=True)
df["redes_sociais"].replace({0: "nunca", 1: "eventualmente", 2: "frequentemente", np.nan: "não informado"}, inplace=True)

df.head()
```

```python
df.to_csv('s3a://projeto-pi-datalake-trusted/sprint3/autismo_pandemia.csv', index=False)
```
