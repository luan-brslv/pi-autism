```python
import pandas as pd
```

```python
df = pd.read_csv("tea_durante_pandemia.csv",sep=";")
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ", "_",regex=True)
df.head()
```

```python
df.columns = df.columns.str.replace("id_", "id")
df.columns = df.columns.str.replace("mother/father_", "pais_autistas")
df.columns = df.columns.str.replace("child_age_", "idade")
df.columns = df.columns.str.replace("child_sex_", "sexo")
df.columns = df.columns.str.replace("asd_diagnosis_", "nivel_autismo")
df.columns = df.columns.str.replace("language_competences_", "nivel_comunicacao")
df.columns = df.columns.str.replace("cognitive_functioning", "funcoes_cognitivas")
df.columns = df.columns.str.replace("language_and_communication_", "status_comunicacao")
df.columns = df.columns.str.replace("emotional_regulation_", "controle_emocional")
df.columns = df.columns.str.replace("social_interaction_", "interacao_social")
df.columns = df.columns.str.replace("stereotypes_", "comportamento_esteriotipado")
df.columns = df.columns.str.replace("behavioral_problems_", "comportamento")
df.columns = df.columns.str.replace("restricted_interests_", "flexibilidade_interesses")
df.columns = df.columns.str.replace("autonomies_", "autonomia")
# df.columns = df.columns.str.replace("frequencia_frequencia_tv", "frequencia_tv")
# df.columns = df.columns.str.replace("frequencia_frequencia_video-games", "frequencia_video-games")
df.columns = df.columns.str.replace("play_with_frinds_online_", "joga_com_amigos_online")
df.columns = df.columns.str.replace("social_networks_", "redes_sociais")

df.head()

```

```python
# df['id'] = df['id'].astype(int)
df = df.astype({"id":'int'}) 
```

```python
df["pais_autistas"].replace({'1': "mae", 2: "pai"}, inplace=True)
df["sexo"].replace({0: "M", 1: "F"}, inplace=True)
df["nivel_autismo"].replace({1: "leve", 2: "moderado", 3: "grave"}, inplace=True)
df["nivel_comunicacao"].replace({0: "sem_comunicacao", 1: "palavras_simples", 2: "frases_curtas", 3: "frases_complexas"}, inplace=True)
df["funcoes_cognitivas"].replace({1: "baixa", 2: "media", 3: "alta"}, inplace=True)
df["status_comunicacao"].replace({1: "melhorou", 2: "estavel", 3: "piorou"}, inplace=True)
df["controle_emocional"].replace({1: "melhorou", 2: "estavel", 3: "piorou"}, inplace=True)
df["interacao_social"].replace({1: "melhorou", 2: "estavel", 3: "piorou"}, inplace=True)
df["comportamento_esteriotipado"].replace({1: "melhorou", 2: "estavel", 3: "piorou"}, inplace=True)
df["comportamento"].replace({1: "melhorou", 2: "estavel", 3: "piorou"}, inplace=True)
df["flexibilidade_interesses"].replace({1: "melhorou", 2: "estavel", 3: "piorou"}, inplace=True)
df["autonomia"].replace({1: "melhorou", 2: "estavel", 3: "piorou"}, inplace=True)
df["tv"].replace({1: "nunca", 2: "eventualmente", 3: "frequentemente"}, inplace=True)
df["video-games"].replace({0: "nunca", 1: "eventualmente", 2: "frequentemente"}, inplace=True)
df["joga_com_amigos_online"].replace({0: "nunca", 1: "eventualmente", 2: "frequentemente"}, inplace=True)
df["redes_sociais"].replace({0: "nunca", 1: "eventualmente", 2: "frequentemente"}, inplace=True)

df.head()

```
